### 54. Write a Python program to get days between two dates. 

```python
def days_diff(s_date,e_date):
    s_date=pd.to_datetime(s_date,format='%Y-%m-%d')
    e_date=pd.to_datetime(e_date,format='%Y-%m-%d')
    return(e_date - s_date).days  
```
